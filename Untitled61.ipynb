```python
import pandas as pd
import sqlite3
```

```python
orders = pd.read_csv("orders.csv")
orders.head()
```

```python
users = pd.read_json("users.json")
users.head()
```

```python
# Create SQLite database
conn = sqlite3.connect("restaurants.db")
cursor = conn.cursor()

# Read SQL file
with open("/content/restaurants (1).sql", "r") as file:
    sql_script = file.read()

# Execute SQL script
cursor.executescript(sql_script)

# Load restaurants table into DataFrame
restaurants = pd.read_sql("SELECT * FROM restaurants", conn)
restaurants.head()
```

```python
final_df = orders.merge(users, on="user_id", how="left") \
                 .merge(restaurants, on="restaurant_id", how="left")

final_df.head()

```

```python
final_df.to_csv("final_food_delivery_dataset.csv", index=False)
```

```python
import pandas as pd

df = pd.read_csv("final_food_delivery_dataset.csv")
df.head()
```

```python
gold_df = df[df['membership'] == 'Gold']

gold_city_revenue = (
    gold_df.groupby('city')['total_amount']
    .sum()
    .sort_values(ascending=False)
)

gold_city_revenue
```

```python
cuisine_aov = (
    df.groupby('cuisine')['total_amount']
    .mean()
    .sort_values(ascending=False)
)

cuisine_aov
```

```python
user_spend = (
    df.groupby('user_id')['total_amount']
    .sum()
)

count_users = user_spend[user_spend > 1000].count()
count_users
```

```python
# Create rating buckets
df['rating_range'] = pd.cut(
    df['rating'],
    bins=[3.0, 3.5, 4.0, 4.5, 5.0],
    labels=['3.03.5', '3.64.0', '4.14.5', '4.65.0']
)

rating_revenue = (
    df.groupby('rating_range', observed=False)['total_amount']
    .sum()
    .sort_values(ascending=False)
)

rating_revenue
```

```python
gold_city_aov = (
    gold_df.groupby('city')['total_amount']
    .mean()
    .sort_values(ascending=False)
)

gold_city_aov
```

```python
restaurant_count = df.groupby('cuisine')['restaurant_id'].nunique()
cuisine_revenue = df.groupby('cuisine')['total_amount'].sum()

summary = pd.DataFrame({
    'restaurant_count': restaurant_count,
    'total_revenue': cuisine_revenue
}).sort_values('restaurant_count')

summary
```

```python
gold_orders = df[df['membership'] == 'Gold'].shape[0]
total_orders = df.shape[0]

percentage = round((gold_orders / total_orders) * 100)
percentage

```

```python
restaurant_stats = (
    df.groupby('restaurant_name_x')
    .agg(
        avg_order_value=('total_amount', 'mean'),
        total_orders=('order_id', 'count')
    )
)

filtered = restaurant_stats[restaurant_stats['total_orders'] < 20] \
            .sort_values('avg_order_value', ascending=False)

filtered.head()
```

```python
combo_revenue = (
    df.groupby(['membership', 'cuisine'])['total_amount']
    .sum()
    .sort_values(ascending=False)
)

combo_revenue
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['quarter'] = df['order_date'].dt.to_period('Q')

quarter_revenue = (
    df.groupby('quarter')['total_amount']
    .sum()
    .sort_values(ascending=False)
)

quarter_revenue
```
